# Sunday Challenge_2 :Employability Quotient Test (EQT) Analysis

```python
import pandas as pd
import numpy as np
import tkinter as tk
import matplotlib.pyplot as plt

xls_list=['EQT Results/'+'A'+str(i+1)+'.xlsx' for i in range(19)]
coln=[('q'+str(i+1)) for i in range(60)] # question number from 1 to 60


def eqt(str,coln):
    '''eqt function which load the excel file and processed it '''
    a=pd.read_excel(str)
    a.drop([len(a)-2,len(a)-1],axis=0,inplace=True)
    colname=['name','inst','start','compl','time','grade']
    eq_name=['numr%','logi%','verb%','itawa%']
    new_coln=list(colname+coln+eq_name)
    a.columns=new_coln
    grade_data=a[coln]
    grade_data=grade_data.replace(to_replace='0.00',value=-0.25)
    grade_data=grade_data.replace(to_replace='-',value='0.00').astype('float')
    grade_data['grade']=grade_data.sum(axis=1)
    grade_data['numr%']=a['numr%'].str[:5].astype('float')
    grade_data['logi%']=a['logi%'].str[:5].astype('float')
    grade_data['verb%']=a['verb%'].str[:5].astype('float')
    grade_data['itawa%']=a['itawa%'].str[:5].astype('float')
    grade_data['inst']=a['inst']
    
    time=a['time'].str.extract(r'((\d+) [hour]+|(\d+) [mins]+ (\d+) [secs]+)|((\d+) [hour]+ (\d+) [mins]+)|(\d+) [mins]')
    time.fillna(0)
    time=time.drop([0,4,5,6],axis=1).fillna(0).astype('int')
    time_data=time[1]*3600+time[2]*60+time[3]+time[7]*60
    grade_data['time(s)']=time_data
    
    grade_data=grade_data[grade_data['time(s)']!=0]
    
    col = grade_data.pop("inst")
    grade_data.insert(0,'inst',col)
    return grade_data

#loading 19 Excel files 
a_lis=[eqt(xls_list[i],coln) for i in range(19)]


#Concatenate all excel files
new_dat=pd.concat(a_lis,ignore_index=True, sort=False)

new_dat['time(s)']=new_dat['time(s)'].where(new_dat['time(s)'] < 3600,3600) # changing the time to 1hr who exceed 1hour

#creating a processed data as csv file 
new_dat.to_csv('EQT.csv')

#feature Engineering 
instn=[('A'+str(i+1)) for i in range(19)]
grad_avg=[round(new_dat[new_dat['inst']==i]['grade'].mean(),2) for i in instn]
numr_avg=[round(new_dat[new_dat['inst']==i]['numr%'].mean(),2) for i in instn]
logi_avg=[round(new_dat[new_dat['inst']==i]['logi%'].mean(),2) for i in instn]
verb_avg=[round(new_dat[new_dat['inst']==i]['verb%'].mean(),2) for i in instn]
itaw_avg=[round(new_dat[new_dat['inst']==i]['itawa%'].mean(),2) for i in instn]
time_avg=[round(new_dat[new_dat['inst']==i]['time(s)'].mean()/60,2) for i in instn]


avg_data={'inst':instn,'numr%':numr_avg,'logi%':logi_avg,'verb%':verb_avg,'itawa%':itaw_avg,'grade':grad_avg,'time(m)':time_avg}


agg_data=pd.DataFrame(avg_data)
agg_data.index=instn

avg_grade=new_dat['grade'].sum()/len(new_dat['grade'])
stu_ab=new_dat[new_dat['grade']>avg_grade]
no_stu_abv=[len(stu_ab[stu_ab['inst']==i]) for i in instn]
no_stu=[len(a_lis[i]) for i in range(19)]
per_stud=np.array(no_stu_abv)/np.array(no_stu)


avg_time=[np.round(stu_ab[stu_ab['inst']==i]['time(s)'].sum(),0) for i in instn]
np.seterr(divide='ignore', invalid='ignore') #ignore the warning on true divide
time_tkn_abv_avg=np.array(avg_time)/np.array(no_stu_abv)

stu_data=pd.DataFrame({'inst':instn,'No.of students':no_stu,'stud_abv_overl_avg':no_stu_abv,'%stud_abv_avg':np.round(per_stud*100,2),'time(m)_abv_avg':np.round(time_tkn_abv_avg/60,0)})
stu_data=stu_data.fillna(0)

# Creating a DATAFrame for plotting etc...
cons_data=agg_data.set_index('inst').join(stu_data.set_index('inst'))
cons_data
cons_data.to_csv('cons_data.csv')
```

# Institute performance grade wise

```python
%matplotlib
br=agg_data[['inst','grade']].plot(kind='bar',figsize=(20,70),rot=360)
br.set_xlabel('Institute',size=20)
br.set_ylabel('grade',size=20)
#br.legend(xl,fontsize='large')
br.set_title('Institute performance grade wise',size=20)

for i in br.patches:
    # get_x pulls left or right; get_height pushes up or down
    br.text(i.get_x()-.02,i.get_height()+.7,str(round(i.get_height(),2)),fontsize=12,color='black') 
```

# Overall EQT Institute wise

```python
x= agg_data[['inst','numr%','logi%','verb%','itawa%']]
y= x.set_index('inst')
z=y.groupby('inst',sort=0).mean()
z.plot.bar(stacked=True,title='Overall EQT Institute wise',rot=360)
plt.ylabel('Percentage')

row_lb=[row for row in cons_data.iloc[:,0:4].values]    
r=np.array(row_lb).transpose()
xpos=0
ypos=0
a=0
j=0
while a<=75:   
    for i in range(len(r)):
        ypos+=r[i][j]
        plt.text(xpos-0.25,ypos-6,str(r[i][j]),color='white')
        a=a+1
    j+=1
    xpos+=1
    ypos=0   
```
